```python
df_people_pool = pd.read_csv(pathdata + 'data_output/Base_pool_individuos_solo_con_replicas_actuales.csv', low_memory=False, index_col=0)
df_houses_pool = pd.read_csv(pathdata + 'data_output/Base_pool_hogares_solo_con_replicas_actuales.csv', low_memory=False, index_col=0)
```

```python
df_temp = df_people_pool.query('CH06 >= 14')
```

```python
df_inactivos = df_temp[(df_temp['P47T_new']!=-9) & (df_temp['ESTADO']==3) & (df_temp['CAT_INAC']!=3)][['PROVINCIA', 'PONDII_new', 'P47T_new', 'CH04']]
```

```python
n_obs = df_inactivos.groupby(['CH04', 'PROVINCIA'])['PONDII_new'].count()
n_pob = df_inactivos.groupby(['CH04', 'PROVINCIA'])['PONDII_new'].sum()
n_sin_ingresos = df_inactivos[df_inactivos['P47T_new']==0].groupby(['CH04', 'PROVINCIA'])['PONDII_new'].sum()
n_con_ingresos = df_inactivos[df_inactivos['P47T_new']>0].groupby(['CH04', 'PROVINCIA'])['PONDII_new'].sum()

tabla_inactivos = pd.concat([n_obs, n_pob, n_sin_ingresos, n_con_ingresos], axis=1)
tabla_inactivos.columns = ['N obs', 'N tot', 'Sin ingresos', 'Con ingresos']

tabla_inactivos = tabla_inactivos.unstack(level=0)
```

```python
tasas = tabla_inactivos[('Con ingresos')] / tabla_inactivos[('N tot')]
```
