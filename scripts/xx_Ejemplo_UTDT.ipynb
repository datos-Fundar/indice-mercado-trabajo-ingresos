```python
numerador = df_temp[df_temp['ESTADO']==2].groupby(['CH04', 'PROVINCIA'])['PONDERA'].sum()       # agrupo por genero y provincia y sumo
    
size = df_temp[df_temp['ESTADO']==2].groupby(['CH04', 'PROVINCIA']).size().to_frame().unstack(level=0)
size.columns = size.columns.get_level_values(1)
```

```python
size
```

```python
denominador = df_temp.groupby(['CH04', 'PROVINCIA'])['PONDERA'].sum()
df_tasa = numerador.div(denominador, fill_value=np.nan)
df_tasa = df_tasa.to_frame().unstack(level=0)
df_tasa.rename(columns={1: 'Varon', 2: 'Mujer'}, inplace=True)
```

```python
df_tasa.columns = df_tasa.columns.droplevel(0)
```

```python
df_tasa
```

```python
df_tasa = df_tasa.join(size)
df_tasa.rename(columns={1: 'N_v', 2: 'N_m'}, inplace=True)
```
